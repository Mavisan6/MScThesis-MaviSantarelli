##  1. Rationale for Corridor Identification

### Why Identify Corridors?
1. High current flow areas represent strong movement potential (McRae et al., 2008).
2. Identifying key corridors can inform conservation actions by showing where connectivity is highest (Keeley et al., 2016).
3. Blue-Green Infrastructure (BGI) planning requires knowing where amphibians are likely to move, so interventions can be targeted effectively (Donati et al., 2022).

### How Do We Define Corridors?
We will test two different approaches:

1. Thresholding High-Current Areas → Extracting the top 10% (90th percentile) of connectivity values.
2. Clustering Connectivity Hotspots → Grouping high-movement zones using spatial clustering algorithms (e.g., DBSCAN or k-means).

## 3. Overlay with Urban Green & Grey Spaces

### **Objective**
This step aims to overlay **amphibian connectivity corridors** with **urban green and grey spaces** to assess potential conflicts and identify areas where Blue-Green Infrastructure (BGI) interventions could improve habitat connectivity.

### **Data Inputs**
| Dataset               | Description  | Source |
|-----------------------|-------------|--------|
| **Circuitscape Connectivity Maps** | Represents the ecological corridors for amphibians, highlighting areas of high connectivity. | Circuitscape Output |
| **OpenStreetMap (OSM) Urban Features** | Includes roads, railways, and buildings that may act as barriers to amphibian movement. | OSM (Overpass API / ArcGIS OSM Layer) |
| **Study Area Boundary** | Defines the extent of the analysis. | Study Area Shapefile |

---

### **Feature Extraction & Rationale**
To understand potential barriers to amphibian movement, we extract the following key features from **OpenStreetMap (OSM):**

### 1️⃣ Urban (Grey) Features
These represent **potential barriers** to amphibian dispersal and habitat connectivity.

- **Roads (`highway=*`)**  
  - Extract major roads that may act as significant barriers to movement.  
  - Includes: `motorway`, `trunk`, `primary`, `secondary`, `tertiary`, `residential`.
  - Rationale: Roads contribute to **habitat fragmentation** and **road mortality** for amphibians.

- **Buildings (`building=*`)**  
  - Includes all built-up structures (`residential`, `commercial`, `industrial`).
  - Rationale: **Impermeable surfaces** restrict movement and reduce available habitat.

- **Railways (`railway=*`)**  
  - Extract **railway lines, subways, and tramways**.
  - Rationale: Similar to roads, railways can act as **barriers** and may require **BGI interventions** (e.g., culverts, green bridges).

- **Industrial & Parking Areas (`landuse=industrial`, `amenity=parking`)**  
  - Extract major industrial areas and large parking zones.
  - Rationale: These areas **lack vegetation**, preventing amphibians from dispersing.

### 2️⃣ Green and Blue Spaces (Potential Habitat & BGI Sites)
These represent **existing and potential corridors** that can support amphibian movement.

- **Parks & Vegetation (`amenities=park`, `landuse=grass`, `landuse=forest`)**  
  - Rationale: Existing **green spaces** provide **habitat patches** and stepping stones for connectivity.

- **Water Bodies & SuDS (`water=wetland`, `water=reservoir`)**  
  - Rationale: **Essential breeding and dispersal sites** for amphibians, aligning with **BGI opportunities**.



### **Methodology in ArcGIS Pro**
### Step 1: Load and Filter OSM Data
1. Download **OSM data** using the **OSM Overpass API** or **ArcGIS OSM Layer**.
2. Use **Select by Attributes** to extract relevant urban features:
   - Roads: `"highway" IN ('motorway', 'trunk', 'primary', 'secondary', 'tertiary', 'residential')`
   - Buildings: `"building" IS NOT NULL`
   - Railways: `"railway" IS NOT NULL`
   - Industrial & Parking: `"landuse" IN ('industrial', 'commercial') OR "amenity" = 'parking'`
3. Clip the extracted features to the **study area boundary**.


### Step 2: Overlay with Amphibian Connectivity
1. **Binarize** connectivity maps:
   - Apply **90th percentile threshold** to retain high-connectivity corridors.
2. Convert binarized corridors into **vector polygons** using **Raster to Polygon**.
3. **Intersect** connectivity polygons with:
   - **Urban features (roads, buildings, railways)**
   - **Green spaces & SuDS (parks, water bodies)**
     
### Step 3: Compute Overlap & Identify High-Conflict Zones
1. **Calculate Area Overlap**:
   - Use **Tabulate Intersection** to compute **% of connectivity area overlapping** each urban feature.
   - Example expected outputs:
     - 🚧 **Road overlap**: 25.4%
     - 🏢 **Building overlap**: 18.2%
     - 🚉 **Railway overlap**: 8.7%
     - 🌳 **Park overlap**: 42.1% (existing connectivity support)

2. **Identify High-Priority BGI Sites**:
   - Locate **high-connectivity corridors** that currently **lack BGI elements**.
   - Prioritize areas for **SuDS ponds, habitat restoration, or ecological corridors**.

### Expected Outputs
✅ **Statistical Summary**  
- % of connectivity area overlapping **urban barriers** (roads, buildings, railways).
- % of connectivity area intersecting **green/blue spaces**.

✅ **Maps & Visualizations**  
- Heatmaps of **high-conflict connectivity areas**.
- Overlay maps showing **opportunities for BGI interventions**.

### Interpretation & Next Steps
- **Assess urban conflict zones** where amphibian corridors are obstructed.
- **Prioritize BGI interventions** where connectivity gaps exist.
- **Integrate findings** into the final recommendations for amphibian-friendly urban planning.


```python
# Import necessary libraries
import geopandas as gpd  # Handles vector data (shapefiles)
import rasterio  # Reads and processes raster data
import rasterio.features  # Converts raster data into vector features
import numpy as np  # Handles numerical operations
import matplotlib.pyplot as plt  # Visualization
from shapely.geometry import shape  # Handles geometric transformations
import pandas as pd  # Handles tabular data
```

```python
# Define file paths
# Define file paths
inv_path = r"C:/GIS_Course/MScThesis-MaviSantarelli/results/Connectivity/Inv_Bufo_Clipped_Results.asc_cum_curmap.asc"
log_path = r"C:/GIS_Course/MScThesis-MaviSantarelli/results/Connectivity/Log_Bufo_Clipped_Results.asc_cum_curmap.asc"

green_areas_shp = r"C:\GIS_Course\MScThesis-MaviSantarelli\data\OSM\GreenAreas.shp"
blue_areas_shp = r"C:\GIS_Course\MScThesis-MaviSantarelli\data\OSM\BlueAreas.shp"
grey_areas_lines_shp = r"C:\GIS_Course\MScThesis-MaviSantarelli\data\OSM\GreyAreas_Lines.shp"
grey_areas_shp = r"C:\GIS_Course\MScThesis-MaviSantarelli\data\OSM\GreyAreas.shp"
study_area_shp = r"C:\GIS_Course\MScThesis-MaviSantarelli\data\StudyArea\Study_Area.shp"
```

```python
# Load shapefiles into GeoDataFrames and reproject them to EPSG:27700
green_areas = gpd.read_file(green_areas_shp).to_crs(epsg=27700)
blue_areas = gpd.read_file(blue_areas_shp).to_crs(epsg=27700)
grey_areas_lines = gpd.read_file(grey_areas_lines_shp).to_crs(epsg=27700)
grey_areas = gpd.read_file(grey_areas_shp).to_crs(epsg=27700)
study_area = gpd.read_file(study_area_shp).to_crs(epsg=27700)

print("All shapefiles successfully reprojected to EPSG:27700.")
```

```python
# Function to process connectivity raster and convert to polygons
def process_connectivity_raster(raster_path, study_area, method_name):
    print(f"Processing {method_name} connectivity raster...")

    with rasterio.open(raster_path) as src:
        connectivity_data = src.read(1)  # Read first band (connectivity values)
        connectivity_transform = src.transform  # Get metadata for spatial reference

    # Compute 90th percentile threshold
    threshold = np.percentile(connectivity_data, 90)
    binary_connectivity = connectivity_data > threshold

    # Convert binary raster to vector polygons
    shapes = list(rasterio.features.shapes(binary_connectivity.astype(np.uint8), transform=connectivity_transform))
    connectivity_polygons = gpd.GeoDataFrame({"geometry": [shape(geom) for geom, _ in shapes]}, crs=study_area.crs)

    # Clip to study area
    connectivity_polygons = gpd.clip(connectivity_polygons, study_area)
    print(f"{method_name} raster binarized and converted to vector polygons.")

    return connectivity_polygons
```

```python
# Process both connectivity maps
inv_connectivity = process_connectivity_raster(inv_path, study_area, "Inverse Reclassification")
log_connectivity = process_connectivity_raster(log_path, study_area, "Logarithmic Reclassification")
```

### **The code above**:
1. Loads the Circuitscape raster.
2. Computes the 90th percentile threshold (top 10% of highest connectivity).
3. Converts raster to vector polygons.
4. Clips the output to study area.

```python
# Function to perform overlay analysis
def overlay_analysis(connectivity_polygons, green_areas, blue_areas, grey_areas, grey_areas_lines, method_name):
    print(f"Performing overlay analysis for {method_name} connectivity...")

    # Intersect connectivity with urban and BGI layers
    green_intersection = gpd.overlay(connectivity_polygons, green_areas, how="intersection")
    blue_intersection = gpd.overlay(connectivity_polygons, blue_areas, how="intersection")
    grey_intersection = gpd.overlay(connectivity_polygons, grey_areas, how="intersection")

    # Convert grey line features to polygons (buffering roads/railways)
    grey_areas_lines["geometry"] = grey_areas_lines.buffer(20)  # 20m buffer
    grey_lines_intersection = gpd.overlay(connectivity_polygons, grey_areas_lines, how="intersection")

    print(f"Overlay completed for {method_name} connectivity.")

    return {
        "green": green_intersection,
        "blue": blue_intersection,
        "grey": grey_intersection,
        "grey_lines": grey_lines_intersection
    }
```

```python
# Perform overlay analysis for both reclassification methods
inv_overlays = overlay_analysis(inv_connectivity, green_areas, blue_areas, grey_areas, grey_areas_lines, "Inverse")
log_overlays = overlay_analysis(log_connectivity, green_areas, blue_areas, grey_areas, grey_areas_lines, "Logarithmic")
```

### **The code above:**
1. Finds overlap between connectivity corridors and:
    * 🚧 Urban infrastructure.
    * 🌳 Green areas.
    * 💧 Water/SuDS.
2. Buffers roads/railways by 20m to approximate impact zone.

```python
# Function to calculate overlap percentages
def calculate_overlap(connectivity_gdf, feature_gdf, feature_name, method_name):
    total_connectivity_area = connectivity_gdf.area.sum()
    feature_overlap_area = feature_gdf.area.sum()
    percentage_overlap = (feature_overlap_area / total_connectivity_area) * 100
    return {"Method": method_name, "Feature": feature_name, "Overlap (%)": round(percentage_overlap, 2)}

# Compute overlap for both methods
overlap_results = []

for method, conn, overlays in [("Inverse", inv_connectivity, inv_overlays), ("Logarithmic", log_connectivity, log_overlays)]:
    for feature, overlay in overlays.items():
        overlap_results.append(calculate_overlap(conn, overlay, feature, method))

# Convert results to DataFrame and display
overlap_df = pd.DataFrame(overlap_results)
# Display overlap results in Jupyter Notebook
from IPython.display import display

# Convert results to DataFrame and display in Jupyter Notebook
overlap_df = pd.DataFrame(overlap_results)

# Show table in Jupyter Notebook
display(overlap_df)

# Save results to CSV for reference
overlap_df.to_csv("Connectivity_Overlap_Comparison.csv", index=False)

print("Overlay analysis and conflict zone identification complete.")
```
