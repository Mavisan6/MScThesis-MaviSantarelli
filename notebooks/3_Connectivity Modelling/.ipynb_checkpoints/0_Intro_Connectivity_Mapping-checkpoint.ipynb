## 1. Introduction

Habitat connectivity is a critical factor influencing species survival in fragmented landscapes, particularly for amphibians that rely on both terrestrial and aquatic habitats for different life stages. Fragmentation due to urbanization and land-use change can create barriers to movement, reducing dispersal opportunities and increasing the risk of local extinctions (Watts et al., 2015; Cayuela et al., 2020). Connectivity modeling provides a means to assess potential movement corridors and identify strategic locations for conservation interventions, such as Blue-Green Infrastructure (BGI), to enhance ecological networks (Donati et al., 2022).

This phase of the study applies circuit theory and resistance surface modeling to evaluate the permeability of the landscape for amphibian movement. Unlike traditional least-cost path analysis, circuit theory models multiple potential pathways simultaneously, providing a more robust representation of movement potential across the study area (McRae et al., 2008).

A key component of connectivity modeling is the creation of resistance surfaces, which assign movement costs to different land cover types based on their suitability for species movement. This study utilizes expert opinion to reclassify land cover data into resistance values, acknowledging the limitations of purely data-driven species distribution models (SDMs) in connectivity assessments.

## 2. Challenges with Species Distribution Models (SDMs) in Connectivity Analysis

Species Distribution Models (SDMs) are widely used to estimate habitat suitability based on species occurrence data and environmental predictors (Guisan & Thuiller, 2005). While SDMs provide valuable insights into potential habitat preferences, they present several limitations when applied to connectivity modeling, many of which were encountered in this study:

1. **Survey Effort Bias**: The SDMs used in this study exhibited a strong bias toward urban areas, likely due to higher species observation rates in more accessible locations. This resulted in inflated habitat suitability predictions near cities and underrepresentation of rural and less-surveyed areas.

2. **Unstable Model Predictions**: Initial SDM results varied significantly depending on the choice of pseudo-absence generation methods. The sensitivity of models to pseudo-absence selection led to inconsistencies in predicted habitat suitability across different model runs, complicating interpretation and application for connectivity analysis.

3. **Limited Reflection of Movement Ecology**: While SDMs may successfully identify potentially suitable habitats, they do not account for barriers or facilitators of movement, such as roads, rivers, and land-use patterns. As amphibians rely on fine-scale landscape features, this omission limited the models' applicability for connectivity assessments.

4. **Multicollinearity in Predictor Variables**: Some environmental predictors used in the SDMs were highly correlated, leading to challenges in model interpretation. Attempts to mitigate collinearity, such as variable selection and standardization, influenced model performance and affected habitat suitability predictions.

## 3. Justification for Using Expert Opinion in Resistance Mapping

Given these challenges, expert opinion was incorporated to develop resistance surfaces for connectivity modeling in Circuitscape. Resistance mapping assigns movement costs to different land cover types based on ecological knowledge rather than species presence data, allowing for a more realistic assessment of potential dispersal corridors and barriers (Zeller et al., 2012; Trainor et al., 2013).

### **Advantages of Expert-Based Resistance Mapping:**

1. **Focuses on Landscape Permeability:** Unlike SDMs, resistance mapping is explicitly designed to model species movement and account for barriers to dispersal.

2. **Incorporates Fine-Scale Landscape Features:** Expert knowledge integrates critical landscape features, such as road mortality risks, vegetation structure, and water bodies, that may not be well captured by broad-scale SDM predictors.

3. **Reduces Dependence on Biased Occurrence Data:** Since SDMs are heavily influenced by occurrence data, which can be geographically biased, expert-driven resistance mapping ensures that connectivity assessments are not skewed by uneven survey efforts.

4. **Supports Connectivity Planning in Conservation:** Many real-world conservation and land management strategies rely on expert-driven resistance models for corridor design and restoration planning (Rayfield et al., 2016).

By integrating expert opinion with GIS-based connectivity modeling, this phase of the study aims to generate resistance surfaces that accurately represent amphibian movement potential across the landscape. The subsequent analysis will apply Circuitscape in Julia to model connectivity patterns, identify ecological corridors, and inform strategic placement of BGI interventions.

## 4. Reclassifying the Land Cover to Resistance Raster

### **Choice of Land Cover Map**

For this study, the **EUNIS (European Nature Information System) habitat classification** has been selected as the base land cover map for resistance mapping. This choice was informed by expert opinion, emphasizing the **comprehensive mapping coverage for Scotland** and its potential for broader application across **mainland Europe**. Using EUNIS ensures consistency in methodology, making it **scalable and transferable** beyond the study area. The standardized classification system allows for **comparative assessments** and facilitates future applications in international conservation planning.

### **Reclassification Process**

The land cover reclassification involves assigning **resistance values** to different EUNIS habitat types based on expert ecological knowledge and species movement constraints. Resistance values represent the difficulty amphibians face when traversing different land cover types, with **low values indicating high permeability** (e.g., wetlands) and **high values indicating strong barriers** (e.g., roads and urban infrastructure).

### 4.2 Literature Review: Methods for Resistance Reclassification

In landscape connectivity modeling, various techniques are used to reclassify habitat suitability into resistance values. The choice of method significantly influences the accuracy of movement predictions for species. Below are commonly used approaches:

#### **1 Inverse Suitability Scaling**
- **Concept:** Resistance is assigned as the inverse of habitat suitability, assuming that highly suitable habitats pose minimal movement resistance while unsuitable habitats act as barriers.
- **Formula:**  
  $$
  \text{Resistance} = 11 - \text{Suitability}
  $$
- **Example Use:** This approach is commonly used in habitat modeling, where expert-derived suitability scores are transformed into resistance surfaces.

#### **2. Logarithmic Transformation**
- **Concept:** Resistance values are scaled logarithmically to reflect a higher differentiation between low and high suitability habitats, emphasizing major landscape barriers.
- **Formula:**  
  $$
  \text{Resistance} = \log_{10}(11 - \text{Suitability})
  $$
- **Example Use:** This method is used to represent species movement constraints, particularly in fragmented landscapes.

#### **3. Exponential Scaling**
- **Concept:** Resistance increases exponentially as suitability decreases, exaggerating the impact of unsuitable areas as barriers.
- **Formula:**  
  $$
  \text{Resistance} = e^{(11 - \text{Suitability})}
  $$
- **Example Use:** This scaling can improve corridor predictions by emphasizing significant barriers.

#### **4. Threshold-Based Reclassification**
- **Concept:** Suitability values are grouped into categorical resistance classes (low, medium, high), simplifying the resistance layer while maintaining key habitat connectivity features.
- **Example Use:** This classification facilitates conservation decision-making by simplifying complex resistance surfaces.

- Ensure uniformity in scaling before applying transformations.

#### **Step 2: Implementing Reclassification Techniques**
- **Approach 1: Inverse Suitability Scaling**
  - Convert suitability scores to resistance using:
    $$
    \text{Resistance} = 11 - \text{Suitability}
    $$
  - Normalize the range (1-10 or 1-1000) depending on modeling requirements.

- **Approach 2: Logarithmic Transformation**
  - Apply:
    $$
    \text{Resistance} = \log_{10}(11 - \text{Suitability})
    $$
  - Adjust output values to ensure appropriate contrast between movement corridors and barriers.

#### **Step 3: Comparing Reclassification Outcomes**
- Visualize resistance maps for each method.
- Conduct statistical analysis to assess differences in resistance distributions.
- Compare resistance layers' impact on connectivity model outputs.

#### **Step 4: Selection of the Best Method**
- Assess model performance using sensitivity analysis.
- Select the most ecologically relevant approach for further connectivity modeling.

## 5. Comparing Square and Hexagonal Raster Grids for Resistance Mapping

The choice of raster grid shape significantly influences the accuracy of ecological modeling, particularly for resistance surface mapping in species connectivity studies. Traditional raster analyses commonly use square grids; however, recent studies suggest that **hexagonal grids** may offer advantages in reducing directional bias and improving spatial analysis accuracy (Birch et al., 2007).

This section explores the benefits and limitations of square versus hexagonal raster grids and proposes an experimental comparison to assess their impact on amphibian connectivity modeling.

### 5.1 Square vs. Hexagonal Raster Grids

#### **1. Square Grids: The Traditional Approach**
Square grids are the most commonly used format in GIS and raster-based ecological modeling due to their simplicity and computational efficiency. Their primary advantages include:

- **Ease of Implementation**: Square grids are well-supported in GIS software and computational models.
- **Hierarchical Scaling**: Square grids allow straightforward aggregation or subdivision of cells for multi-resolution analysis (Birch et al., 2007).
- **Alignment with Built Environments**: Many anthropogenic features, such as roads and land-use boundaries, align naturally with square grids.

However, square grids also introduce **directional bias** due to their four primary neighboring cells, where diagonal connectivity distances are longer than horizontal or vertical distances (Tucker et al., 2018).

#### **2. Hexagonal Grids: A More Natural Representation**
Hexagonal grids have gained popularity in ecological modeling due to their ability to better represent natural movement patterns and spatial relationships. Their advantages include:

- **Equal Neighboring Distances**: Each hexagonal cell has six equidistant neighbors, preventing bias in connectivity calculations (Birch et al., 2007).
- **Reduced Edge Effects**: The lower perimeter-to-area ratio of hexagons minimizes boundary artifacts, leading to smoother ecological transitions (Tucker et al., 2018).
- **More Accurate Sampling**: Studies suggest that hexagonal grids improve spatial sampling efficiency and reduce error in interpolated datasets.
- **Improved Representation of Natural Habitats**: Hexagons better capture curved or irregularly shaped habitat patches, aligning more closely with species dispersal patterns.

Despite these benefits, hexagonal grids **require specialized processing** and are not as widely supported in raster-based GIS software as square grids.

### 5.2 Proposed Experiment: Comparing Grid Structures

To assess the impact of **grid structure** on resistance surface modeling, this study will compare square and hexagonal raster grids using the **EUNIS land cover resistance reclassification** dataset. The key steps are:

1. **Generate Resistance Surfaces**  
   - Convert the expert-assigned suitability scores into resistance values.
   - Implement two reclassification techniques: **inverse suitability scaling** and **logarithmic transformation**.
  
2. **Create Square and Hexagonal Grids**  
   - Generate two raster grids: **one using a traditional square grid** and **one using a hexagonal grid of equivalent resolution**.
   - Ensure that the total area covered remains consistent across both grids.

3. **Run Circuitscape Connectivity Modeling**  
   - Apply the resistance layers to Circuitscape.
   - Compare the connectivity pathways generated from square and hexagonal grids.

4. **Statistical & Visual Analysis**  
   - Analyze connectivity differences between square and hexagonal rasters.
   - Assess computational efficiency, corridor identification accuracy, and sensitivity to resistance reclassification.


### 5.3 Expected Outcomes

- Improved understanding of **grid structure effects** on connectivity modeling.
- Identification of whether **hexagonal grids provide a more ecologically relevant representation** of amphibian movement.
- A reproducible methodology for integrating **alternative grid structures** into GIS-based resistance mapping.
